Carico il dataset Iris

```python
from sklearn.datasets import load_iris
import pandas

iris = load_iris()
df = pandas.DataFrame(data=iris.data, columns=iris.feature_names)
df["target"] = pandas.Series(iris.target)
```

Standardizzo le variabili numeriche

```python
df.dtypes
```

```python
variabili_numeriche = list(df.columns[0:-1])
variabili_numeriche
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[variabili_numeriche])
scaled_data
```

Eseguo la PCA e mostro la varianza spiegata

```python
import matplotlib.pyplot as pyplot

pca = PCA().fit(scaled_data)

pyplot.plot(range(1, pca.n_components_ + 1), pca.explained_variance_ratio_, marker='o')
```
